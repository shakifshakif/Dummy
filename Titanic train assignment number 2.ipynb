```python
import pandas as pd
import numpy as np
import matplotlib as mat
import matplotlib.pyplot as plt

titanic = pd.read_excel(r'C:\Users\shakif.rauf\Shakif_Python_Anaconda codes and data science\titanic.xlsx')
titanic.head()
```

```python
#age group wise distribution of non survivors

dead = titanic.loc[titanic['Survived']=='No']
dead.head()

xax = dead['Age_Group'].unique()
yax = dead['Age_Group'].value_counts()
plt.bar(xax,yax,color='blue',edgecolor = 'black',width=.1)
plt.title('Age Group wise non survivors',fontsize = 12)
plt.ylabel('No of non survivors')
plt.xlabel('Age Group',fontsize = 12)
plt.ylim(0,300)

plt.bar(xax,yax)
ax = plt.axes()
for i,j in zip(xax,yax):
    ax.annotate(str(j)+('  Dead'),xy=(i,j))
```

```python
dead_male =  dead.loc[dead['Sex'] == 'M',:]
dead_female = dead.loc[dead['Sex'] == 'F',:]

plt.figure(figsize=(7,6))
xax = dead['Age_Group'].unique()
yax_m = (dead_male['Age_Group'].value_counts())
yax_f = (dead_female['Age_Group'].value_counts())
plt.bar(xax,yax_m,color='blue',edgecolor = 'black',width = .5)
plt.bar(xax,yax_f,color='pink',edgecolor = 'black',bottom=yax_m,width = .5)
plt.title('Age Group wise non survivors',fontsize = 12)
plt.ylabel('No of non survivors')
plt.xlabel('Age Group',fontsize = 12)
plt.ylim(0,300)
plt.show()
```
