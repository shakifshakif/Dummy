```python
import pandas as pd
import numpy as np
import matplotlib as mat
import matplotlib.pyplot as plt

titanic = pd.read_excel(r'C:\Users\shakif.rauf\Shakif_Python_Anaconda codes and data science\titanic.xlsx')
titanic.head()
```

```python
dead = titanic.loc[titanic['Survived']=='No']
dead.head()

val = dead['Sex'].value_counts()
label = set(dead['Sex'])
pie = plt.pie(val,explode = (0,0.1),labels = label,autopct = '%.0f%%',colors = ['red','orange'])
plt.axis('equal')
plt.title('Death of Male and Female',fontsize = 15)
plt.legend(labels = label,loc = 2)
pie
```
